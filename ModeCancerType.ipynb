```
import warnings
warnings.filterwarnings('ignore')
```

```
from fastai import *
from fastai.vision import *
```

```
densenet=load_learner('/content/drive/My Drive/Models_MultiClass','densenet121.pkl')
resnet=load_learner('/content/drive/My Drive/Models_MultiClass','resnet50.pkl')
vgg16=load_learner('/content/drive/My Drive/Models_MultiClass','vgg16bn.pkl')
```

```
lister=[]
for files in os.listdir('/content/Testing'):
    lister.append(files)
```

```
lister=['Actinic keratosis','Basal cell carcinoma','Benign keratosis','Dermatofibroma','Melanoma','Melanocytic nevus','Vascular lesion']
```

```
import os 
import cv2
acc1=acc2=acc3=0
n_total=0
for count,files in enumerate(os.listdir('/content/Testing')):
    print(files)
    path=os.path.join('/content/Testing',files)
    for subfiles in os.listdir(path):
        #print(subfiles)
        pathi=os.path.join(path,subfiles)
        img=open_image(pathi)
        tens1=densenet.predict(img)[-1].numpy()
        tens2=resnet.predict(img)[-1].numpy()
        tens3=vgg16.predict(img)[-1].numpy()
        
        c1=lister[np.argmax(tens1)]
        c2=lister[np.argmax(tens2)]
        c3=lister[np.argmax(tens3)]
        n_total+=1
        if c1==files:
            acc1+=1
        if c2==files:
            acc2+=1
        if c3==files:
            acc3+=1
print(acc1,acc2,acc3)
        
```
